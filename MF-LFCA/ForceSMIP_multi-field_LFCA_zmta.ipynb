High-frequency variability (e.g., ENSO) is filtered out.

### Data structure, training and evaluation data

LFCA analyzes a data matrix with a size of the number of timesteps (n<sub>t</sub>) by the number of spatial points (n<sub>lat</sub> x n<sub>lon</sub>). The notebook is set up to use monthly surface air temperature values, though you can change the variable used or try modifying it to use annual means. 

<p align="center">
  <img width=800 src="figs/anomaly_maps.png" />
</p>

This data matrix will be created for the **training data** and the **evaluation data**. For the training data, we have many ensemble members, and we can compute the forced response as the ensemble average. For the evaluation data, we only have one realization from each model, just as we only have one realization of the real world. The challenge of ForceSMIP is to come up with methods that can approximate the forced response from these single realizations, removing internal variability as we would if we were able to take an ensemble mean.

### LFCA

The basis for LFCA is principal component analysis (aka EOF analysis). From this we find a set of basis functions (EOFs) that efficiently describe the variability in the data matrix. A key parameter in LFCA is the number of EOFs to include, which is labelled 'trucation' in the code. LFCA then looks for linear combinations of the included EOFs that maximize the ratio of low-frequency to total variance, by comparing the (principal component) amplitude of the patterns in a low-pass filtered version of the data matrix compared to their amplitude in the full data. Here, we use a 10-year (120-month) low-pass filter, but this can be modified. 

This analysis is schematically depicted below. The result are what we call low-frequency patterns (LFPs) and their low-frequency component (LFC) timeseries. 

<p align="center">
  <img width=800 src="figs/LFCA_flowchart.png" />
</p>

### Forced response estimate (LFP filtering)

To reconstruct an estimate of the forced response, we multiply LFP*LFC for a few of the leading LFP/LFC combinations to once again make a matrix with dimensions of the number of timesteps (n<sub>t</sub>) by the number of spatial points (n<sub>lat</sub> x n<sub>lon</sub>). Wills et al. (2020) calls this LFP filtering. The number of LFP/LFC combinations to include, which is called `N_LFCs` in the code, is the other key hyperparameter of the analysis.

### Utilizing the training data

The training data is used for choosing the hyperparamters `trucation` and `N_LFCs`. Here, the notebook simply tries a few different values of each, then compares the LFP filtered forced reponse estimate to the "right answer", which for the case of the training data is the ensemble mean. This notebook shows a few different possible skill estimates, and the user is free to choose which values work best for their needs depending on what they want to optimize, i.e., anomaly correlation coefficient (ACC) vs. root-mean square error (RMSE), spatial pattern of trends vs. spatiotemporal variance at monthly timescales.

### Define LFCA, Eigenvalue Analysis, and helper functions

Look under the hood. Here are the functions needed for LFCA. 

Note that this is currently setup to use numpy. There is an option to switch it to xarray, which is more efficient, but it requires fixing a bug in the installed eofs package: https://stackoverflow.com/questions/71740621/eofs-xarray-raising-typeerror-using-a-dataarray-to-construct-a-variable-is-ambi. 

```python
"""
New xarray implementation of Low-Frequency Component Analysis (LFCA)

Author: Oliver Mehling (University of Torino)
Based on Numpy Version by: Zhaoyi Shen (Caltech)
Minor edits by: Robb Jnglin Wills (ETH Zurich)

Reference:  Wills et al.: Disentangling Global Warming, Multidecadal Variability,
            and El Niño in Pacific Temperatures,
            Geophysical Research Letters 45, 2487-2496 (2018)
"""
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from eofs.standard import Eof
from scipy.signal import convolve, butter, filtfilt

### Helper functions ###
def low_pass_weights(window, cutoff):
    """Calculate weights for a low pass Lanczos filter.

    Args:

    window: int
        The length of the filter window.

    cutoff: float
        The cutoff frequency in inverse time steps.
        
    References
    ----------
        from https://scitools.org.uk/iris/docs/v1.2/examples/graphics/SOI_filtering.html

    """
    order = ((window - 1) // 2 ) + 1
    nwts = 2 * order + 1
    w = np.zeros([nwts])
    n = nwts // 2
    w[n] = 2 * cutoff
    k = np.arange(1., n)
    sigma = np.sin(np.pi * k / n) * n / (np.pi * k)
    firstfactor = np.sin(2. * np.pi * cutoff * k) / (np.pi * k)
    w[n-1:0:-1] = firstfactor * sigma
    w[n+1:-1] = firstfactor * sigma
    return w[1:-1]

def filter_padding(ts, window, ftype='mirror', detrend=True, detrend_poly=1): 
    ts_pad = np.zeros(2*window+len(ts))
    if detrend:
        t = np.arange(len(ts))
        z = np.polyfit(t,ts,detrend_poly)
        p = np.poly1d(z)
        ts_in = ts-p(t)
    else:
        ts_in = ts
    ts_pad[window:-window] = ts_in[:]
    if ftype == 'mirror':
        ts_pad[:window] = ts_in[:window][::-1]
        ts_pad[-window:] = ts_in[-window:][::-1]
    elif ftype == 'periodic':
        ts_pad[:window] = ts_in[-window:]
        ts_pad[-window:] = ts_in[:window]
    else:
        raise ValueError('in filter_padding: ftype must be one of "mirror" or "periodic".')
    if detrend:
        t_pad = np.arange(-window,len(ts)+window)
        ts_pad = ts_pad+p(t_pad)
    return ts_pad

def filter_ts(ts, cutoff, filter_type='lanczos', padding_type='mirror', detrend=True, detrend_poly=1): 
    lanczos_weights=low_pass_weights(cutoff*2+1,1./cutoff) # weights for Lanczos filter
    n_pad=int(np.ceil(len(lanczos_weights)/2))
    
    # Padding
    ts_mirr=filter_padding(ts,n_pad,padding_type,detrend=detrend,detrend_poly=detrend_poly)
    
    # Filtering
    if filter_type=='lanczos':
        # Lanczos filter
        return convolve(ts_mirr,lanczos_weights,'same')[n_pad:-n_pad]
    elif filter_type=='butter':
        # Butterworth filter
        # TODO: here, the cutoff frequency needs to be doubled
        # to obtain the same result as for Lanczos - why?
        b,a = butter(3,1./(cutoff/2),btype='low')
        return filtfilt(b,a,ts_mirr)[n_pad:-n_pad]
    else:
        raise ValueError('in filter_ts: filter_type must be one of "lanczos" or "butter".')
        
### Main LFCA function ###
def lfca(x, cutoff, truncation, weights, **kwargs):
    if x.ndim!=2:
        raise ValueError('x must have dimension 2 for LFCA')
    
    # Scale vector from weights
    scale = np.sqrt(np.transpose(weights)/np.sum(weights))
    
    # center data
    x = x - np.mean(x,axis=0)
    xs = x * scale.T

    
    # Compute EOFs using eofs package
    # xarray version: eofs_xr=Eof(xs, center=False, ddof=1)
    eofs_xr=Eof(xs, center=False, ddof=1)
    fvar = eofs_xr.varianceFraction()
    cvar = np.cumsum(fvar)
    
    if truncation<1:
        trunc = [ n for n,i in enumerate(cvar) if i>truncation ][0]
    else:
        trunc = truncation
    print("Truncation = " + str(trunc))
    
    # Principal component time series (PC_k)
    pcs=eofs_xr.pcs(npcs=trunc, pcscaling=1)
    
    
    # Filtering of PCs
    pcs_filt=np.zeros(pcs.shape)
    for i in range(trunc):
        pci = pcs[:,i]
        pci_filt = filter_ts(pci, cutoff, **kwargs)
        pcs_filt[:,i]=pci_filt[:]
        #print(np.std(pci),np.std(pci_filt))
    
    # Compute low-frequency components
    cov_lowfreq=np.cov(pcs_filt,rowvar=False)
    eig_lowfreq, eigv_lowfreq = np.linalg.eigh(cov_lowfreq) # Eigenvalues r_k, Eigenvectors e_k
    eig_argsort = np.argsort(eig_lowfreq)[::-1].copy() # Guarantee that eigenvalues are sorted in descending order
    eig_lowfreq = eig_lowfreq[eig_argsort].copy()
    eigv_lowfreq = eigv_lowfreq[:,eig_argsort].copy()

    uvec=eofs_xr.eofs(neofs=trunc, eofscaling=1).T@eigv_lowfreq # u_k
    # xarray version: lfcs=xs@uvec # Low-frequency components (LFC_k)
    lfcs = np.dot(xs, uvec)
    lfps=eofs_xr.eofs(neofs=trunc, eofscaling=2).T@eigv_lowfreq # v_k = Low-frequency patterns (LFP_k)
    # NB: Alternative formula (lfps=xs.T@lfcs) gives the right pattern, but wrong scaling
    
    # Choose positive sign of LFCs/LFPs
    for j in range(lfps.shape[1]):
        if np.dot(lfps[:,j], scale.flatten())<0:
            lfps[:,j] = -lfps[:,j]
            lfcs[:,j] = -lfcs[:,j]
    lfps=lfps/scale # Re-scale LFPs (non-weighted)
    lfps = np.transpose(lfps)
    
    return lfcs, lfps
```

### Define a function to read in data

We're going to loop over many models and realizations for training and evaluation data. To make this more readable and to reduce repeating code, we are going to define a function to do this operation.

```python
def load_realization(fn, vid, time_period, reference_period):
    """
    load_realization(fn, vid, time_period, reference_period)
    
    Function loads in data for a given file, fn, and variable, vid. It
    selects data for a given time_period and calculates the anomalies
    relative to a user-defined reference_period. The function returns arrays
    of the dimensions (time, lat, lon), the 3D anomaly map, and the global
    mean time series. 
    
    Inputs:
    -------
    fn (str) : filename
    vid (str) : variable id
    time_period (tuple(str, str)) : tuple of the start and end of the time period
                                    e.g., ("1900-01-01", "1949-12-31")
    reference_period (tuple(str, str)) : tuple of the start and end of the reference period
                                         used to calculate anomalies e.g., ("1900-01-01", "1949-12-31")
                                         
    Returns:
    --------
    ts_3d (xr.DataArray) : monthly average anomaly values [time, lat, lon]
    ts_gm (xr.DataArray) : monthly average, global mean anomaly values
    """
    # open dataset
    ds = xc.open_dataset(fn)
    # if specified, subset training/validation data to specific period
    if tv_time_period is not None:
        ds = ds.sel(time=slice(time_period[0], time_period[1]))
    # get departures
    ds = ds.temporal.departures(vid, freq="month", reference_period=reference_period)
    # If you wanted annual averages instead, you could use the following:
    # ds = ds.temporal.group_average(vid, freq="year", weighted=False)
    ts_3d = ds[vid]
    # clean up 
    ds.close()
    # return values
    return ts_3d
```

### Read in evaluation data

We start with the evaluation data, because this data includes observations...and missing data. We will load the evaluation data and create a missing data mask, which we will apply to the training/evaluation data as well. That way the training and evaluation data will be consistent.

```python
def load_eval_data(ncvar): # The function is not self-contained. Variables like the tier and root dir need to be defined globaly.    
    # first we search for the evaluation data
    epath = "/".join([root_dir, "Evaluation-" + eval_tier, cmipTable[ncvar], ncvar])
    efiles = glob.glob(epath + "/*.nc")
    efiles = sorted(efiles, key=natsort)

    # initialize dictionary to store data
    evaluation_anomaly_maps = {}
    missing_data_mask = {}
    vid = cmipVar[ncvar]

    # loop over evaluation files
    for im, fn in enumerate(efiles):
        # get evaluation identifier
        model = fn.split("/")[-1].split("_")[2].split(".")[0]
        # print progress
        print(str(im + 1) + " / " + str(len(efiles)) + ": " + model)
        # read in data for realization
        ts_3d = load_realization(fn, vid, tv_time_period, reference_period)
        # store anomaly map
        evaluation_anomaly_maps[model] = ts_3d

        # create mask for missing data
        tmp = np.mean(ts_3d, axis=0)
        missing_data_mask[model] = np.where(np.isnan(tmp), np.nan, 1)
    
    return evaluation_anomaly_maps, missing_data_mask
```

### Loop over training models and retrieve monthly anomaly maps

```python
def load_training_data(ncvar):
    # initialize dictionary to store data
    anomaly_maps = {}
    anomaly_map_emean = {}
    missing_data_mask_training = {}
    all_members = {}
    vid = cmipVar[ncvar]
    # loop over models
    models = training_models
    for im, model in enumerate(models):
        # start timer
        stime = clocktime.time()
        # initialize nested dictionary for model data
        anomaly_maps[model] = {}
        anomaly_map_emean[model] = 0
        missing_data_mask_training[model] = {}
        all_members[model] = {}
        # get model files
        mpath = "/".join([root_dir, "Training", cmipTable[ncvar], ncvar, model])
        mfiles = glob.glob(mpath + "/*.nc")
        # parse file names to get list of model members
        # CESM2 has a non-CMIP naming convention
        if model == "CESM2":
            members = [p.split("ssp370_")[-1].split(".1880")[0] for p in mfiles]
        else:
            members = [p.split("_")[-1].split(".")[0] for p in mfiles]
        all_members[model] = sorted(members, key=natsort)
        # print progress
        print(str(im + 1) + " / " + str(len(models)) + ": " + model + " (" + str(len(members)) + " members)")
        # loop over model members
        for imm, member in enumerate(all_members[model]):
            # define member filename
            fn = glob.glob(mpath + "/*_" + member + ".*.nc")
            # make sure filename is unique
            if len(fn) != 1:
                raise ValueError("Unexpected number of model members")
            else:
                fn = fn[0]
            # load data for realization
            ts_3d = load_realization(fn, vid, tv_time_period, reference_period)
            # store first N anomaly maps from training models
            # also store all data for the validation models
            if ((imm < n_members) & (model in training_models)):# | (model in validation_models):
                # store masked data in array
                anomaly_maps[model][member] = ts_3d
            # ensemble mean anomaly map
            anomaly_map_emean[model] = anomaly_map_emean[model] + ts_3d/len(members)
            # update progress
            print(".", end="")
        # create mask for missing data
        tmp = np.mean(anomaly_map_emean[model].values, axis=0)
        missing_data_mask_training[model] = np.where(np.isnan(tmp), np.nan, 1)
        # print time elapse for model
        etime = clocktime.time()
        print()
        print("Time elapsed: " + str(etime - stime) + " seconds")
        print()
    return anomaly_maps, anomaly_map_emean, missing_data_mask_training, all_members
```

## Flattening & Normalizing data

We are now reshaping our data to have the dimensions lat*lon x time. Further, because we are working with multiple different variables and units, we are normalizing each variable and model-member by the trace of its covariance matrix. 

```python
def flatten_normalize_training_data(ncvar):
    # Creating dictionaries for the normalized anomalie maps & the trace of the covariance matrix which is used for normalization
    anomaly_maps_norm = {}
    traces = {}
    
    for model in training_models:
        traces[model] = {}
        anomaly_maps_norm[model] = {}
        print(model)
        for member in all_members[model][:n_members]:
            traces[model][member] = {}
            anomaly_maps_norm[model][member] = {}
            # Getting the current anomalie map and reshaping it to be lat*lon x time
            if ncvar == "zmta":
                current_data = np.squeeze(all_anomaly_maps[ncvar][model][member].stack(shape=['lat','plev']))
            else:
                current_data = all_anomaly_maps[ncvar][model][member].stack(shape=['lat','lon'])
            # Calculating the Covariance matrix (We copy the first maxrix to avoid an [unnecessary] error)
            cov_matrix = np.matmul(np.transpose(current_data.values.copy()), current_data.values)

            # Adding up each non-nan element along the diagonal of the covariance matrix
            trace = 0
            for i in range(0, np.shape(cov_matrix)[0]):
                if np.isnan(cov_matrix[i,i]) == False:
                    trace = trace + cov_matrix[i,i]

            # Saving trace so we can later convert back to real units
            traces[model][member] = trace
            # Saving the normalized data
            anomaly_maps_norm[model][member] = current_data / trace
    return anomaly_maps_norm, traces
```

```python
all_anomaly_maps_norm = {}
all_traces = {}

for ncvar in ncvars:
    all_anomaly_maps_norm[ncvar], all_traces[ncvar] = flatten_normalize_training_data(ncvar)
```

## LFCA of training members

```python
var_idx = 1

if var_idx == 0 :
    xdim = "lon"
    xcoord = 'lon'
    x_axis = lon_axis
    nx = nlon
else:
    xdim = "plev"
    xcoord = 'plev'
    x_axis = plev_axis
    nx = nlev

# time axis
time_plot = [t.year + t.month/12 for t in time.values]

# intialize skill metrics
mse = np.zeros((n_models,n_members,len(truncations),len(N_LFCs)))
corr = np.zeros((n_models,n_members,len(truncations),len(N_LFCs)))
mse_trend = np.zeros((n_models,n_members,len(truncations),len(N_LFCs)))
corr_trend = np.zeros((n_models,n_members,len(truncations),len(N_LFCs)))
mse_raw = np.zeros((n_models,n_members))
corr_raw = np.zeros((n_models,n_members))
mse_trend_raw = np.zeros((n_models,n_members))
corr_trend_raw = np.zeros((n_models,n_members))

cosw = np.sqrt(np.cos(lat_axis*np.pi/180))
cosw = cosw/np.mean(cosw)

lfps_all_separated = {}
for ncvar in ncvars:
    lfps_all_separated[ncvar] = {}

# loop over training models/members, truncation, and N_LFCs to evaluate the LFP filtering estimate of the forced response against the ensemble mean
ig=0
for model in training_models:#enumerate(all_members):
    print(model)
    # start timer
    stime = clocktime.time()
    # ensemble mean for each model
    X_emeans = []
    for ncvar in ncvars:
        lfps_all_separated[ncvar] = {}
        lfps_all_separated[ncvar][model] = {}
    X_emean = all_anomaly_map_emeans[ncvars[var_idx]][model].values 
    X_emeans.append(X_emean)

    if var_idx == 0 :
        X_emean = xr.DataArray(np.squeeze(X_emeans), coords=[time_plot,lat_axis,x_axis], dims=["time","lat",xdim])
    else:
        X_emean = xr.DataArray(np.transpose(np.squeeze(X_emean), (0, 2, 1)), coords=[time_plot,lat_axis,x_axis], dims=["time","lat",xdim])
    im = 0
    for member in all_members[model][:n_members]:
        X_total_all = []
        for ncvar in ncvars:
            lfps_all_separated[ncvar][model][member] = {}
        X_total = all_anomaly_maps[ncvars[var_idx]][model][member].values
        X_total_all.append(X_total)
        if var_idx == 0 :
            X_total = xr.DataArray(np.squeeze(X_total_all), coords=[time_plot,lat_axis,x_axis], dims=["time","lat",xdim])
        else:
            X_total = xr.DataArray(np.transpose(np.squeeze(X_total), (0, 2, 1)), coords=[time_plot,lat_axis,x_axis], dims=["time","lat",xdim])
        
        # Statistics for spatiotemporal variance
        sigma_total = np.sqrt((X_total*X_total*cosw).mean(xcoord).mean('lat').mean('time').values)
        sigma_emean = np.sqrt((X_emean*X_emean*cosw).mean(xcoord).mean('lat').mean('time').values)
        corr_raw[ig,im] = ((X_total*X_emean*cosw).mean(xcoord).mean('lat').mean('time').values/(sigma_total*sigma_emean)) 
        diff = X_total-X_emean
        mse_raw[ig,im] = (diff*diff*cosw).mean(xcoord).mean('lat').mean('time').values
        # Compute linear trend
        X_total_trend = X_total.polyfit('time',1).isel(degree=0).polyfit_coefficients
        X_emean_trend = X_emean.polyfit('time',1).isel(degree=0).polyfit_coefficients
        # Pattern MSE (including global mean difference)
        diff = X_total_trend-X_emean_trend
        mse_trend_raw[ig,im] = (diff*diff*cosw).mean(xcoord).mean('lat').values
        # Zero mean
        X_total_trend = X_total_trend - (X_total_trend*cosw).mean(xcoord).mean('lat')
        X_emean_trend = X_emean_trend - (X_emean_trend*cosw).mean(xcoord).mean('lat')
        # Statistics for trend
        sigma_total_trend = np.sqrt((X_total_trend*X_total_trend*cosw).mean(xcoord).mean('lat').values)
        sigma_emean_trend = np.sqrt((X_emean_trend*X_emean_trend*cosw).mean(xcoord).mean('lat').values)
        corr_trend_raw[ig,im] = (X_total_trend*X_emean_trend*cosw).mean(xcoord).mean('lat').values/(sigma_total_trend*sigma_emean_trend)
        it = 0
        for truncation in truncations:
            inl = 0
            for N_LFC in N_LFCs: 
                # extract relevant LFCs and LFPs from dictionary
                lfcs = lfcs_all[model][member][str(truncation)]
                lfps = lfps_all[model][member][str(truncation)]
                # put back in missing data 
                icol_ret = icol_ret_all[model]
                nins = nfeatures - len(icol_ret[0]) #np.size(icol_disc[0])
                nrows = lfps.shape[0]
                lfps_aug = np.zeros((nrows,lfps.shape[1]+nins))
                lfps_aug[:] = np.nan
                lfps_aug[:,icol_ret[0]] = lfps
                # forced response estimate from LFP filtering
                
                # Separating the lfps for each variable & converting back to real units
                current_trace = all_traces[ncvars[var_idx]][model][member]
                if var_idx == 0 :
                    lfps_all_separated[ncvars[var_idx]][model][member][str(truncation)] = lfps_aug[:, 0:(nlat*nx)] * current_trace
                    X_forced = np.matmul(lfcs[:,0:N_LFC], (lfps_aug[0:N_LFC, 0:(nlat*nx)] *current_trace))
                    X_forced = X_forced.reshape(n_timesteps,nlat,nx)
                else:
                    lfps_all_separated[ncvars[var_idx]][model][member][str(truncation)] = lfps_aug[:, (nlat*nlon):(nlat*nlon+nlat*nx)] * current_trace
                    X_forced = np.matmul(lfcs[:,0:N_LFC], (lfps_aug[0:N_LFC, (nlat*nlon):(nlat*nlon+nlat*nx)] *current_trace))
                    X_forced = X_forced.reshape(n_timesteps,nlat,nx)
                if var_idx == 0 :
                    X_forced = xr.DataArray(X_forced, coords=[time_plot,lat_axis,lon_axis], dims=["time","lat",xdim])
                else:
                    X_forced = xr.DataArray(X_forced, coords=[time_plot,lat_axis,x_axis], dims=["time","lat",xdim])
                # Statistics for spatiotemporal variance
                sigma_forced = np.sqrt((X_forced*X_forced*cosw).mean(xcoord).mean('lat').mean('time').values)
                sigma_emean = np.sqrt((X_emean*X_emean*cosw).mean(xcoord).mean('lat').mean('time').values)
                corr[ig,im,it,inl] = (X_forced*X_emean*cosw).mean(xcoord).mean('lat').mean('time').values/(sigma_forced*sigma_emean)
                diff = X_forced-X_emean
                mse[ig,im,it,inl] = (diff*diff*cosw).mean(xcoord).mean('lat').mean('time').values
                # Compute linear trend
                X_forced_trend = X_forced.polyfit('time',1).isel(degree=0).polyfit_coefficients
                X_emean_trend = X_emean.polyfit('time',1).isel(degree=0).polyfit_coefficients
                # Pattern MSE (including global mean difference)
                diff = X_forced_trend-X_emean_trend
                mse_trend[ig,im,it,inl] = (diff*diff*cosw).mean(xcoord).mean('lat').values
                # Zero mean
                X_forced_trend = X_forced_trend - (X_forced_trend*cosw).mean(xcoord).mean('lat')
                X_emean_trend = X_emean_trend - (X_emean_trend*cosw).mean(xcoord).mean('lat')
                # Statistics for trend
                sigma_forced_trend = np.sqrt((X_forced_trend*X_forced_trend*cosw).mean(xcoord).mean('lat').values)
                sigma_emean_trend = np.sqrt((X_emean_trend*X_emean_trend*cosw).mean(xcoord).mean('lat').values)
                corr_trend[ig,im,it,inl] = (X_forced_trend*X_emean_trend*cosw).mean(xcoord).mean('lat').values/(sigma_forced_trend*sigma_emean_trend)
                inl = inl +1
            it = it +1
        im = im +1
        print(im)
    ig = ig +1
    etime = clocktime.time()
    print()
    print("Time elapsed: " + str(etime - stime) + " seconds")
    print()
```

Optimal params based on graphs: 15 - 30 EOFS more EOFs reduce variance but can induce bias lower RMSE lower pattern correlation & 1 LFC & longer time period, more LFCs should be included. Huge change in ACC, i.e. skill from individual member preds.

```python
def flatten_normalize_eval_data(ncvar):
    # Creating dictionaries for the normalized anomalie maps & the trace of the covariance matrix which is used for normalization
    anomaly_maps_norm = {}
    traces = {}
    
    for model in eval_models:
        # Getting the current anomalie map and reshaping it to be lat*lon x time
        if ncvar == "zmta":
            current_data = np.squeeze(all_evaluation_anomaly_maps[ncvar][model].stack(shape=['lat','plev']))
        else:
            current_data = all_evaluation_anomaly_maps[ncvar][model].stack(shape=['lat','lon'])
        # Calculating the Covariance matrix
        cov_matrix = np.matmul(np.transpose(current_data.values.copy()), current_data.values)
        
        # Adding up each non-nan element along the diagonal of the covariance matrix
        trace = 0
        for i in range(0, np.shape(cov_matrix)[0]):
            if np.isnan(cov_matrix[i,i]) == False:
                trace = trace + cov_matrix[i,i]
        # Saving trace so we can later convert back to real units
        traces[model] = trace
        # Saving the normalized data
        anomaly_maps_norm[model] = current_data / trace
    return anomaly_maps_norm, traces
```

```python
# manually set hyperparameters (based on training)
truncation = 0.8
N_LFC = 1

# initialize dictionaries for LFCA output
lfps_eval_all_combi = {}
lfcs_eval_all_combi = {}
X_forced_eval_all = {}
X_total_eval_all = {}
lfps_eval_all_separated = {}
for ncvar in ncvars:
    lfps_eval_all_separated[ncvar] = {}
    for eval_model in eval_models:
        lfps_eval_all_separated[ncvar][eval_model] = {}        

ie = 0
for eval_model in eval_models:
    print("Model " + eval_model)
    
    lfcs_eval_all_combi[eval_model] = {}
    lfps_eval_all_combi[eval_model] = {}
    X_forced_eval_all[eval_model] = {}
    X_total_eval_all[eval_model] = {}

    lon_axis = all_evaluation_anomaly_maps["tos"][eval_model].lon
    plev_axis = all_evaluation_anomaly_maps["zmta"][eval_model].plev
    lat_axis = all_evaluation_anomaly_maps["tos"][eval_model].lat
    time = all_evaluation_anomaly_maps["tos"][eval_model].time
    y, x = np.meshgrid(lat_axis,lon_axis)
    
    current_data = []
    all_area_weights = []
    for ncvar in ncvars:
        # this is the data from each training model/member
        sst_anomalies =  all_eval_anomaly_maps_norm[ncvar][eval_model]
        current_data.append(sst_anomalies)
        # convert missing data mask from nans to zeros
        mask = np.nan_to_num(all_missing_data_masks[ncvar][eval_model], nan=0)
        # create cos(lat) weighting, with mask for missing data 
        if ncvar == "zmta":
            y, x = np.meshgrid(lat_axis,plev_axis)
            area = np.cos(y*np.pi/180.) * np.squeeze(mask)
            area_weights = np.reshape(area,(nlat * nlev,1),order='F')
        else:
            y, x = np.meshgrid(lat_axis,lon_axis)
            area = np.cos(y*np.pi/180.) * np.transpose(mask)
            area_weights = np.reshape(area,(nlat * nlon,1),order='F')
        all_area_weights.append(area_weights)
        
    # dimensions of the data matrix
    n_timesteps = len(time)
    if ncvars[1] == "zmta":
        nfeatures = nlat * (nlon + nlev)
    else:
        nfeatures = nlat * nlon * len(ncvars)
    
    # concatenating area weights
    area_weights = np.concatenate(all_area_weights)
    print(np.shape(area_weights))    
    # Model "1I" for precipitation has a different calendar than the other models. This is fixed here.
    if(eval_model == "1I" and ( any("pr" in ncvars for ncvar in ncvars)==True or any("monmaxpr" in ncvars for ncvar in ncvars)==True ) and len(ncvars) > 1):
        non_pr_index = np.where((np.array(ncvars) != "pr") & (np.array(ncvars) != "monmaxpr"))[0]
        pr_index = [x for x in np.linspace(0, len(ncvars)-1, len(ncvars), dtype=np.int64)if x not in non_pr_index]
        for i in range(len(pr_index)):
            current_data[pr_index[i]]["time"] = current_data[non_pr_index[0]].time
    # Model "1F" for daily variables has a different calendar than the other models. This is fixed here.
    if(eval_model == "1F" and ( any("monmintasmin" in ncvars for ncvar in ncvars)==True or any("monmaxtasmax" in ncvars for ncvar in ncvars)==True or any("monmaxpr" in ncvars for ncvar in ncvars)==True ) and len(ncvars) > 1):
        non_daily_index = np.where((np.array(ncvars) != "monmintasmin") & (np.array(ncvars) != "monmaxtasmax") & (np.array(ncvars) != "monmaxpr"))[0]
        daily_index = [x for x in np.linspace(0, len(ncvars)-1, len(ncvars), dtype=np.int64)if x not in non_daily_index]
        for i in range(len(daily_index)):
            current_data[daily_index[i]]["time"] = current_data[non_daily_index[0]].time

    x_all = np.concatenate((current_data[0].values, current_data[1].values), axis=1)
    area_weights[np.where(np.isnan(np.mean(x_all,0)))] = 0
    #x_all = xr.concat(current_data, dim="shape")
        
    # these give the spatial points (columns) where there are not and are missing data, in icol_ret and icol_disc, respectively
    icol_ret = np.where(area_weights!=0)
    icol_disc = np.where(area_weights==0)
    # remove missing spatial points from x and area_weights
    x = x_all[:,icol_ret[0]]
    area_weights = np.transpose(area_weights[icol_ret[0],:])
    

    # perform LFCA for each evaluation model 
    lfcs, lfps =  lfca(x, cutoff, truncation, area_weights)

    # put back in missing data 
    nins = len(icol_disc[0])
    nrows = lfps.shape[0]
    lfps_aug = np.zeros((nrows,lfps.shape[1]+nins))
    lfps_aug[:] = np.nan
    lfps_aug[:,icol_ret[0]] = lfps
    
    # time axis
    time_plot = [t.year + t.month/12 for t in time.values]
    
    all_X_forced = []
    for i in range(0, len(ncvars)):
        # Separating the lfps for each variable and converting them back to real units
        current_trace = all_eval_traces[ncvars[i]][eval_model]
        lfps_eval_all_separated[ncvars[i]][eval_model] = lfps_aug[:, i*(nlat*nlon):(i+1)*(nlat*nlon)] * current_trace
        X_forced = np.matmul(lfcs[:,0:N_LFC],(lfps_aug[0:N_LFC, i*(nlat*nlon):(i+1)*(nlat*nlon)] * current_trace))
        if ncvars[i]=='zmta' :
            X_forced = X_forced.reshape(n_timesteps,nlat,nlev)
            current_data[i] = xr.DataArray(current_data[i].to_numpy().reshape(n_timesteps,nlat,nlev) * current_trace , coords=[time_plot,lat_axis,plev_axis], dims=["time","lat","plev"])
            X_forced_eval_all[eval_model][ncvars[i]] = xr.DataArray(X_forced, coords=[time_plot,lat_axis,plev_axis], dims=["time","lat","plev"])
        else:
            X_forced = X_forced.reshape(n_timesteps,nlat,nlon)
            current_data[i] = xr.DataArray(current_data[i].to_numpy().reshape(n_timesteps,nlat,nlon) * current_trace , coords=[time_plot,lat_axis,lon_axis], dims=["time","lat","lon"])
            X_forced_eval_all[eval_model][ncvars[i]] = xr.DataArray(X_forced, coords=[time_plot,lat_axis,lon_axis], dims=["time","lat","lon"])
    
    # put output in library
    lfcs_eval_all_combi[eval_model] = lfcs
    lfps_eval_all_combi[eval_model] = lfps_aug
    X_total_eval_all[eval_model] = current_data
    
    ie = ie + 1
```

## Visualizing LFP Filtering for Tropical Upper-Tropospheric Temperature

The real benefits of LFCA aren't necessarily for the trend pattern over the full time period but in getting information about how this forced response pattern evolves in time (i.e., the full spatiotemporal complexity of the forced response). We visualize what LFP filtering is doing to the raw data for the tropical upper tropospheric temperature

```python
# Here you have to specifiy the index of the variable you want to plot. I.e. var_idx = 0 is the first variable in your ncvars array, var_idx = 1, the second etc.
var_index = 1

time_plot = [t.year + t.month/12 for t in time.values]
f = plt.figure(figsize=(11, 16))
for ie, eval_model in enumerate(eval_models):
    plt.subplot(5, 2, ie+1)
    X_forced = X_forced_eval_all[eval_model][ncvars[var_index]]
    X_total = X_total_eval_all[eval_model][var_index]
    TUT_forced = (X_forced).sel(plev=slice(25000,15000),lat=slice(-15,15)).mean('plev').mean('lat')
    TUT_total = (X_total).sel(plev=slice(25000,15000),lat=slice(-15,15)).mean('plev').mean('lat')
    plt.plot(time_plot,TUT_total)
    plt.plot(time_plot,TUT_forced)
    plt.title(eval_model+' Tropical Upper Troposphere')
    plt.legend(['Raw','LFP Filtered'])
plt.tight_layout()
```
